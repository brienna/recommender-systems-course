**Top movies for someone who has seen Toy Story.**

Rank movies by highest association score with Toy Story. In other words, for each movie, calculate the percentage of Toy Story raters who also rated that movie. 

*association score = [(movie AND Toy Story) / Toy Story]*

```python
# Get data subset for Toy Story ratings that are not null, also dropping unnecessary columns
data_subset = data[data['1: Toy Story (1995)'].notnull()].drop(columns=['User', 'Gender (1 =F, 0=M)']).reset_index(drop=True)
data_subset
```

```python
percentages = ((data_subset.count()) / data_subset['1: Toy Story (1995)'].count())
percentages
```

```python
# WOULD NEED TO SPLIT OUT THE NUMBERS TO BE ABLE TO SORT BY THEM TOO, 
# SINCE THIS IS CURRENTLY LOOKING AT THE MOVIE TITLE AS A WHOLE
percentages.sort_index().sort_values(kind='mergesort', ascending=False)
```

**Mean rating difference by gender.** 

Recompute the mean rating for each movie separately for males and for females. Calculate the overall mean rating (across all ratings) for males and females. Find the two movies that have the greatest differences (one where men are most above women, and one where women are most above men) along with the differences in average. Also find the difference in overall rating averages (female average - male average).

```python
# Calculate the overall mean rating for males and females 
# (AND NOTE THAT IN THE INDEX, 0=MALE, 1=FEMALE)
means = data.groupby(by='Gender (1 =F, 0=M)').mean().drop(columns=['User']).reset_index(drop=True)
means
```

```python
# Find differences in average ratings, with ratings from men subtracted from those from women 
# (an arbritrary choice, can subtract women from men using periods=-1 argument)
diffs = means.diff() 
diffs
```

```python
# Find the movie that women rated the most higher than men 
# (maximum difference, as we subtracted men from women)
diffs.columns[diffs.eq(diffs.iloc[1].max()).any()][0]
```

```python
# What was its difference in average rating?
diffs.iloc[1].max()
```

```python
# Find the movie that men rated the most higher than women
# (minimum difference, as we subtracted men from women)
diffs.columns[diffs.eq(diffs.iloc[1].min()).any()][0]
```

```python
# What was its difference in average rating?
abs(diffs.iloc[1].min())
```

```python
# Find difference in the overall rating averages (female average - male average)
data.groupby(by='Gender (1 =F, 0=M)').sum().drop(columns=['User']).reset_index(drop=True).mean(axis=1).diff()[1]
```

**% of ratings 4+ (liking), by gender.**


Compute the % of ratings 4+ separately for males and females. Find the two movies with the largest difference in each direction, as above. Indicate whether men or women are more likely to rate movies 4 stars or above.

```python
gender_counts = data.groupby(by='Gender (1 =F, 0=M)').count()
gender_counts
```

```python
# Rank % of ratings 4+ for females
women_likes = ((data[data['Gender (1 =F, 0=M)'] == 1] >= 4.0).sum() / gender_counts.iloc[1]).sort_values(ascending=False)
women_likes
```

```python
# Rank % of ratings 4+ for males
men_likes = ((data[data['Gender (1 =F, 0=M)'] == 0] >= 4.0).sum() / gender_counts.iloc[0]).sort_values(ascending=False)
men_likes
```

```python
# Sort movies with greatest female-male percentage liking gap
(women_likes - men_likes).sort_values(ascending=False)
```

```python
# Sort movies with the greatest male-female percentage liking gap
(men_likes - women_likes).sort_values(ascending=False)
```
